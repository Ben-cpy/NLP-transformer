## 目录
- [transformer_intro.ipynb](transformer_intro.ipynb) — 环境检查与基础可视化
- [transformer_core.ipynb](transformer_core.ipynb) — 模型核心组件
- [transformer_training.ipynb](transformer_training.ipynb) — 训练与复制任务示例
- [transformer_analysis.ipynb](transformer_analysis.ipynb) — 模型分析与可视化
- [transformer_exercises.ipynb](transformer_exercises.ipynb) — 实验与练习
- 公用模块: [transformer_utils.py](transformer_utils.py)

## 使用建议
1. 先打开 intro 了解可视化接口。
2. 在 core 阅读组件签名，必要时跳转到 transformer_utils.py 查看实现。
3. 在 training 运行复制任务并生成 copy_task_model.pt。
4. 在 analysis 载入模型查看Attention与层输出。
5. 在 exercises 完成扩展实验并记录结果。
